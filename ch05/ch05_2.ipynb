```python
noiseimg = gImg.copy()
```

```python
from random import randint
```

```python
# 10% noise
noiseRatio = 20
whiteMax = noiseRatio/2
blackMax = noiseRatio
```

```python
noiseimg = gImg.copy()

for h in range(0, gImg.shape[0]):
  for w in range(0, gImg.shape[1]):
    val = randint(0, 100)
    if(val < whiteMax):
      noiseimg[h, w] = 255
    elif(val < blackMax):
      noiseimg[h, w]= 0

cv2_imshow(noiseimg)
```

```python
outImg =cv2.medianBlur(noiseimg,3)
cv2_imshow(outImg)
```

```python
outImg2 =cv2.medianBlur(noiseimg,5)
cv2_imshow(outImg2)
```
