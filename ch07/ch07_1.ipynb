```python
img = cv2.imread('/content/drive/MyDrive/colabp/Images/shrinking.bmp')

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_BGR2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python
#outImg=cv2.resize(gImg, (0,0), fx=0.3, fy=0.3,interpolation=cv2.INTER_NEAREST)
#outImg=cv2.resize(gImg, (0,0), fx=0.5, fy=0.5,interpolation=cv2.INTER_LINEAR)
outImg=cv2.resize(gImg, (0,0), fx=0.3, fy=0.3,interpolation=cv2.INTER_CUBIC)
cv2_imshow(outImg)
```
