```python
#img = cv2.imread('/content/drive/MyDrive/colabp/Images/abcdef.bmp')
img = cv2.imread('/content/drive/MyDrive/colabp/Images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)

gImg1 = cv2.cvtColor(RGB_img, cv2.COLOR_BGR2GRAY)
print(gImg1.shape)
plt.imshow(gImg1, cmap='gray')
```

```python
gOutImg= np.zeros((gImg1.shape[0],gImg1.shape[1]))
```

```python
for h in range(0,gImg1.shape[0]):
  for w in range(0,gImg1.shape[1]):
    val = gImg1[h,w] % 64
    val=val*4
    gOutImg[h,w] = val

plt.imshow(gOutImg,cmap='gray')
#cv2_imshow(gOutImg)
```

```python
equalval =200
width =5


min = equalval -width
max = equalval +width
```

```python
for h in range(0,gImg1.shape[0]):
  for w in range(0,gImg1.shape[1]):
    if(gImg1[h,w]>min and gImg1[h,w]<max):
      gOutImg[h,w] = 0;
    else:
      gOutImg[h,w] = gImg1[h,w];
plt.imshow(gOutImg,cmap='gray')
```

```python
for h in range(0,gImg1.shape[0]):
  for w in range(0,gImg1.shape[1]):
    val = 255*(gImg1[h,w]/128-1)**2
    gOutImg[h,w] = val
cv2_imshow(gOutImg)
```

```python
planeNumber =1
planeValue = 1

for i in range(planeNumber-1):
  planeValue= planeValue *2
```

```python
for h in range(0,gImg1.shape[0]):
  for w in range(0,gImg1.shape[1]):
    imval= (int)(gImg1[h,w] / planeValue)
    if(imval %2 == 0):
      gOutImg[h,w] = 0
    else:
      gOutImg[h,w] = 255
cv2_imshow(gOutImg)
```
