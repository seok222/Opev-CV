```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/drive/')
```

```python
img = cv2.imread('/content/drive/MyDrive/colabp/Images/scanimg.png')

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_BGR2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python
kernel = np.ones((4,1), np.uint8)
erodeimg = cv2.erode(gImg, kernel, iterations = 1)
cv2_imshow(erodeimg)
```

```python
img = cv2.imread('/content/drive/MyDrive/colabp/Images/fce(salt_pepper noise).bmp')

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_BGR2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python

kernel = np.ones((3,3), np.uint8)
dilateimg = cv2.dilate(gImg, kernel, iterations = 1)
cv2_imshow(dilateimg)
```

```python

kernel = np.ones((3,3), np.uint8)
erodeimg = cv2.erode(dilateimg, kernel, iterations = 2)
cv2_imshow(erodeimg)
```
