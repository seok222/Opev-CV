```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/drive/')
```

```python
img = cv2.imread('/content/drive/MyDrive/colabp/Images/girl.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
plt.imshow(gimg, cmap='gray')
```

```python
gPlus = np.zeros((gimg.shape[0], gimg.shape[1]), dtype = np.ubyte)
```

```python
val = 100

for h in range(gimg.shape[0]):
  for w in range(gimg.shape[1]):
    plusVal = gimg[h,w] + val
    if(plusVal > 255):
      plusVal = 255
    gPlus[h,w] = plusVal

plt.imshow(gPlus, cmap = 'gray')
```
